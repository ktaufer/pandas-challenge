```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file)
```

```python
playercount=len(purchase_data["SN"].unique())
print(f"Total Players = {playercount}")
```

```python
#Purchasing Analysis
averageprice = purchase_data["Price"].mean()
uniqueitems = len(purchase_data["Item ID"].unique())
totalpurchases=len(purchase_data["Purchase ID"])
totalrevenue=purchase_data["Price"].sum()

aveprice=round(averageprice, 2)

print(f"Total revenue from in-game purchases:  {totalrevenue}")
print(f"Total number of purchases:  {totalpurchases}")
print(f"Average item price:  {aveprice}")
print(f"Number of unique items purchased:  {uniqueitems}")
```

```python
newdf=purchase_data[["SN","Price","Gender"]]
grouped=newdf.groupby(["SN"])
regrouped=grouped["Price"].sum()
```

```python
#Gender Demographics
uniqueplayers=purchase_data.drop_duplicates(['SN'])

males= len(uniqueplayers.loc[uniqueplayers["Gender"]=="Male"])
females=len(uniqueplayers.loc[uniqueplayers["Gender"]=="Female"])
others=len(uniqueplayers.loc[uniqueplayers["Gender"]=="Other / Non-Disclosed"])

malepercent = males/len(uniqueplayers)*100
femalepercent = females/len(uniqueplayers)*100
otherpercent = others/len(uniqueplayers)*100

rmalepercent = round(malepercent,2)
rfemalepercent = round(femalepercent, 2)
rotherpercent = round(otherpercent, 2)

print("Player Demographics:  ")
print('----------------------------------------------------------')
print(f"Male players:  {rmalepercent}%, ({males}).")
print(f"Female players:  {rfemalepercent}%, ({females}).")
print(f"Players of undeclared gender:  {rotherpercent}%, ({others}).")
```

```python
#Purchase demographics by gender
# Average Purchase Total per Person by Gender

malerows=purchase_data.loc[purchase_data["Gender"]=="Male"]
femalerows=purchase_data.loc[purchase_data["Gender"]=="Female"]
otherrows=purchase_data.loc[purchase_data["Gender"]=="Other / Non-Disclosed"]

#male1=grouped.loc[grouped["Gender"]=="Male"]
#female1=grouped.loc[grouped["Gender"]=="Female"]
#other1=grouped.loc[grouped["Gender"]=="Other / Non-Disclosed"]

malecount=len(malerows)
femalecount=len(femalerows)
othercount=len(otherrows)

maleave=malerows['Price'].mean()
femaleave=femalerows['Price'].mean()
otherave=otherrows['Price'].mean()

malepurchases=malerows['Price'].sum()
femalepurchases=femalerows['Price'].sum()
otherpurchases=otherrows['Price'].sum()

rmaleave=round(maleave, 2)
rfemaleave=round(femaleave,2)
rotherave=round(otherave,2)

#maleper=round(male1["Price"].mean(),2)
#femaleper=round(female1["Price"].mean(),2)
#otherper=round(other1["Price"].mean(),2)

print("Purchasing Analysis by Gender:")
print("-------------------------------------------------")
print("Males:")
print(f"Number of purchases: {malecount}") 
print(f"Average purchase price:  {rmaleave}")
print(f"Total purchase value:  {malepurchases}")
#print(f"Average purchase per gamer:  {maleper}")

print("-------------------------------------------------")
print("Females:")
print(f"Number of purchases: {femalecount}") 
print(f"Average purchase price:  {rfemaleave}")
print(f"Total purchase value:  {femalepurchases}")
#print(f"Average purchase per gamer:  {femaleper}")

print('-------------------------------------------------')
print('Other/Non-Disclosed Gender:')
print(f"Number of purchases: {othercount}") 
print(f"Average purchase price:  {rotherave}")
print(f"Total purchase value:  {otherpurchases}")
#print(f"Average purchase per gamer:  {otherper}")
```
