```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file = "HeroesofPymoli/Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file)
```

```python
playercount=len(purchase_data["SN"].unique())
print(f"Total Players = {playercount}")
```

```python
#Purchasing Analysis
averageprice = round(purchase_data["Price"].mean(),2)
uniqueitems = len(purchase_data["Item ID"].unique())
totalpurchases=len(purchase_data["Purchase ID"])
totalrevenue=purchase_data["Price"].sum()

purchases=pd.DataFrame([totalrevenue,totalpurchases,averageprice,uniqueitems],
                       ["Total Revenue","Total Number of Purchases","Average Item Price","Number of Unique Items Purchased"])

print("Summary Analysis")
purchases
```

```python
#Gender Demographics
uniqueplayers=purchase_data.drop_duplicates(['SN'])

males= len(uniqueplayers.loc[uniqueplayers["Gender"]=="Male"])
females=len(uniqueplayers.loc[uniqueplayers["Gender"]=="Female"])
others=len(uniqueplayers.loc[uniqueplayers["Gender"]=="Other / Non-Disclosed"])

rmalepercent = round(males/len(uniqueplayers)*100,1)
rfemalepercent = round(females/len(uniqueplayers)*100,1)
rotherpercent = round(others/len(uniqueplayers)*100,1)

demographics=pd.DataFrame({"Gender":["Male","Female","Other/Non-Disclosed"],
                          "Percent of Gamers":[rmalepercent,rfemalepercent,rotherpercent],
                          "Number of Gamers":[males,females,others]})
print("Player Demographics:  ")
demographics
```

```python
#Purchase demographics by gender

malerows=purchase_data.loc[purchase_data["Gender"]=="Male"]
femalerows=purchase_data.loc[purchase_data["Gender"]=="Female"]
otherrows=purchase_data.loc[purchase_data["Gender"]=="Other / Non-Disclosed"]

malecount=len(malerows)
femalecount=len(femalerows)
othercount=len(otherrows)

rmaleave=round(malerows['Price'].mean(),2)
rfemaleave=round(femalerows['Price'].mean(),2)
rotherave=round(otherrows['Price'].mean(),2)

malepurchases=malerows['Price'].sum()
femalepurchases=femalerows['Price'].sum()
otherpurchases=otherrows['Price'].sum()

mr1=malerows.groupby(["SN"]).sum()
fr1=femalerows.groupby(["SN"]).sum()
or1=otherrows.groupby(["SN"]).sum()

mr2=round(mr1["Price"].mean(),2)
fr2=round(fr1["Price"].mean(),2)
or2=round(or1["Price"].mean(),2)

genderpurchases=pd.DataFrame({"Gender":["Males","Females","Other/Non-Disclosed"],
                              "Number of Purchases":[malecount, femalecount,othercount],
                              "Average Purchase":[rmaleave,rfemaleave,rotherave],
                              "Total Purchases":[malepurchases, femalepurchases, otherpurchases],
                              "Average Purchase Per Gamer":[mr2, fr2, or2]})
                              
print("Purchasing Analysis by Gender:")

genderpurchases
```
